```python
import pandas as pd
import os
from sqlalchemy import create_engine
from google.colab import files

# This command will unzip the uploaded file.
print("Unzipping Data.zip...")
!unzip Data.zip

# --- Step 2: Create the database connection engine ---
# This creates a connection to a new or existing SQLite database file.
try:
    engine = create_engine('sqlite:///inventory.db')
    print("Database engine created successfully.")
except Exception as e:
    print(f"Error creating database engine: {e}")
    # Exit if the database connection fails.
    exit()

# --- Step 3: Loop through the files and ingest them in chunks ---
# Set the path to your unzipped data folder.
data_folder_path = '/content/Data'

# Check if the directory exists before proceeding.
if not os.path.exists(data_folder_path):
    print(f"Error: The directory '{data_folder_path}' does not exist.")
else:
    # Get a list of all files in the directory.
    for filename in os.listdir(data_folder_path):
        # Check if the file is a CSV file.
        if filename.endswith('.csv'):
            full_path = os.path.join(data_folder_path, filename)
            table_name = filename[:-4]

            print(f"\nProcessing file: {filename}...")

            # Use a flag to handle the first chunk correctly.
            # The first chunk should replace the table if it exists.
            # Subsequent chunks should append to it.
            first_chunk = True

            try:
                # Read the CSV file in chunks using the 'chunksize' parameter.
                # A chunk size of 100,000 is a good starting point, but you can
                # adjust it based on your file size and available RAM.
                chunk_iterator = pd.read_csv(full_path, chunksize=100000)

                for chunk in chunk_iterator:
                    # Ingest the chunk into the database.
                    if first_chunk:
                        chunk.to_sql(table_name, con=engine, if_exists='replace', index=False)
                        first_chunk = False # Set flag to False after the first chunk
                    else:
                        chunk.to_sql(table_name, con=engine, if_exists='append', index=False)

                    print(f" - Successfully ingested a chunk into '{table_name}'.")

                print(f"File '{filename}' successfully ingested in its entirety.")
            except Exception as e:
                print(f"An error occurred while processing {filename}: {e}")



```
